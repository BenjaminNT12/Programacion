```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('/home/nicolas/Github/Programacion/machileLearning/deeplearning.mplstyle')

```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

train = np.array([x_train, y_train])

print(f"estas son las entradas del entrenamiento: {x_train}")
print(f"estas son salidas del entrenamiento: {y_train}")
```

```python
m = x_train.shape[0] # con shape[0] se obtiene el tamaño de la primera dimension
# print(train)
print(f"este es el tamaño del conjunto de entrenamiento: {m}")
```

```python
i = 1
x_i = x_train[i]
y_i = y_train[i]

print(f"x^{i}, y^{i} = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing prices")
plt.xlabel("Size in square meters")
plt.ylabel("Price in 1000$")
plt.show()
```

```python
w = 200
b = 100

print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i]+b

    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
plt.plot(x_train, tmp_f_wb, c="b", label="Actual Values")

plt.scatter(x_train, y_train, marker="x", c="r", label='Actual Values')

plt.title("Housin prices")

plt.ylabel("Price in 1000s of dollars")

plt.xlabel("Size (1000sqft)")

plt.legend()
plt.show()
```
